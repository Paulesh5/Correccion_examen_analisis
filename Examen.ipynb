```python
#2.- Transformación de cada elemento del csv en json y carga en couchdb mediante código
```

```python
import json
```

```python
import csv
```

```python
csvfile = open('imdb_movies.csv', 'r', encoding='utf-8')
jsonfile = open('imdb_movies.json', 'w')

leer = csv.DictReader(csvfile)

json_movies = []

for i in leer:
    json_movies.append(i)
    
json.dump(json_movies, jsonfile)

jsonfile.close()
csvfile.close()
```

```python
import couchdb
import json

couch = couchdb.Server('http://admin:admin@localhost:5984')
db = couch['movies']

with open('imdb_movies.json', 'r') as jsonfile:
    json_carga = json.load(jsonfile)

for doc in json_carga:
    db.save(doc)
```
